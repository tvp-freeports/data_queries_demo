## 2. Import, complete and save the dataset for the front end analysis

```python
## Imports ##

holdings = pd.read_csv("freeports_data/batch_out_2024/investments.csv")
holdings_deforastation = pd.read_csv("freeports_data/batch_out_BR_SOY_DEFOREST_3K_2024/investments.csv")    
list_mapping=pd.read_csv(".id/company_info.csv")
```

```python
## Merging holdings with company_info ##

list_mapping.rename(columns={"Company name": "Company"}, inplace=True)
flag_cols = [
    'Company legal name',
    'OHCHR25/UN/2025-09-26',
    'OHCHR23/UN/2023-06-30',
    'UNSR_OPT/UN/2025-01-01',
    'DBIO/NGO/2024-01-01',
    'TEST/FREEPORTS/2025-01-01'
]

holdings = holdings.merge(list_mapping[['Company'] + flag_cols], on='Company', how='left')
holdings.head()
```

```python
# Quick environmental list adaptation and merge to the holdings dataframe

holdings['BR_SOY_DEFOREST_3K/TRASE.EARTH/2022-01-01'] = 0

holdings_deforastation['BR_SOY_DEFOREST_3K/TRASE.EARTH/2022-01-01'] = 1
columns_to_add = ['OHCHR25/UN/2025-09-26',
    'OHCHR23/UN/2023-06-30',
    'UNSR_OPT/UN/2025-01-01',
    'DBIO/NGO/2024-01-01',
    'TEST/FREEPORTS/2025-01-01']
holdings_deforastation[columns_to_add] = 0

holdings_deforastation['Company legal name'] = holdings_deforastation['Company']

# Align columns of holdings_deforastation to match the order of holdings
desired_cols = holdings.columns.tolist()
# Reorder to match holdings
holdings_deforastation = holdings_deforastation[desired_cols]
holdings_deforastation.head()

# Append
holdings = pd.concat([holdings, holdings_deforastation], ignore_index=True)
holdings.tail()
```

```python
## Missing merge between holdings and funds_info ##

# 'Theoretical' addings:

# 1. Umbrella fund;
# 3. Management Company;
# 4. Investment Manager;
# 5. Associated banking group;
# 6. Market value --> Market value in EUR;

# .4 Rough-demo version

holdings['Associated banking group'] = np.where(holdings['Format'] == 'DANSKEINVEST-EN24', 'Danske Bank', 'Unicredit')
```

```python
## Quick currency->EUR conversion ##

# import
exchange_rates = pd.read_csv(".id/eurofxref-hist.csv")
exchange_rates.head()

# prep
exchange_rates['EUR']=1.0
exchange_rates=exchange_rates[exchange_rates['Date']=='2024-12-31']

fx_long = exchange_rates.melt(
    id_vars='Date',              
    var_name='Currency',         
    value_name='Exchange_rate' 
)

merged =  pd.merge(holdings, fx_long[['Currency', 'Exchange_rate']], on='Currency', how ='left')

#Conversion
merged['Market value in EUR']=merged['Market value'] / merged['Exchange_rate']
merged['Market value in EUR'] = merged['Market value in EUR'].round(2)
merged.head()
```

```python
## Readability ##

holdings = merged.copy()

# Columns renaming

names = {
    'Document':'Publication date',
    'Subfund': 'Sub-fund',
    'Report page': 'Page',
    'Company': 'search name',
    'Company legal name': 'Investee company legal name',
}

holdings.rename(columns=names, inplace=True)
holdings.head()

# Columns order

order = [
    'Publication date',
    'Sub-fund',
    'Format',
    'Page',
    'search name',
    'Investee company legal name',
    'Financial instrument',
    'Nominal/Quantity',
    'Market value',
    'Market value in EUR',
    'Associated banking group',
    'OHCHR25/UN/2025-09-26',
    'OHCHR23/UN/2023-06-30',
    'UNSR_OPT/UN/2025-01-01',
    'DBIO/NGO/2024-01-01',
    'BR_SOY_DEFOREST_3K/TRASE.EARTH/2022-01-01',
    'TEST/FREEPORTS/2025-01-01'
]

holdings = holdings[order]
holdings.head()
```
